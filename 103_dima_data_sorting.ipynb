```python
train_data = []
with concurrent.futures.ProcessPoolExecutor(max_workers=10) as executor:
    str = DATA_FS103 + '/train' + SUF_ANN + '/*.json'
#   json_files = glob.glob('/home/dima/DSML/UMAMI/data/FoodSeg103/train/ann/*.json')
    json_files = glob.glob(str)
    for jf in json_files:
        suB_cls = []
        suP_cls = []
        with open(jf) as f:
            data = json.load(f)
            fname = f.name
            fheight = data['size']['height']
            fwidth = data['size']['width']
            objects = data['objects']
            for item in objects:
                suB_cls.append(item['classTitle'])
                suP_cls.append(item['tags'])
        train_data.append([fname, fheight, fwidth, suB_cls])



```
