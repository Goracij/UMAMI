```python
############################################################################################
# FUNCTIONS, DEFFINITIONS
############################################################################################

class CRelu(tf.keras.Layer):
    def __init__(self, axis=-1, **kwargs):
        self.axis = axis 
        super(CRelu, self).__init__(**kwargs)

    def build(self, input_shape):
        super(CRelu, self).build(input_shape)

    def call(self, x):
        x = tf.nn.crelu(x, axis=self.axis)
        return x

    def compute_output_shape(self, input_shape):
        output_shape = list(input_shape)
        output_shape[-1] = output_shape[-1] * 2
        output_shape = tuple(output_shape)
        return output_shape

    def get_config(self, input_shape):
        config = {'axis': self.axis, }
        base_config = super(CReLU, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))
```

```python
############################################################################################
# DEFINING THE MODEL
############################################################################################
tf.keras.backend.clear_session()
     
inputed = tf.keras.Input(shape=(256,256,3))


# 0 block (256):
x = tf.keras.layers.Resizing(height = 256,width = 256,interpolation='bilinear',crop_to_aspect_ratio=False,pad_to_aspect_ratio=True,fill_mode='constant',fill_value=0.0,data_format=None)(inputed)
x = tf.keras.layers.Rescaling(1./255)(x)
x = tf.keras.layers.Normalization()(x)
x = tf.keras.layers.Conv2D(filters=128,kernel_size = (7,7),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)

# 1 block (256):
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (7,7),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)


# 2 block (256-128):
y = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-1')(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (7,7),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (7,7),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-2')(x)
x = tf.keras.layers.Conv2D(filters=128,kernel_size = (7,7),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Conv2D(filters=128,kernel_size = (7,7),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Add()([x,y])

# 2 block (128-64):
y = tf.keras.layers.Conv2D(filters=256,kernel_size = (7,7),strides=(2, 2),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (7,7),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (7,7),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-4')(x)
x = tf.keras.layers.Conv2D(filters=256,kernel_size = (7,7),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Conv2D(filters=256,kernel_size = (7,7),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Add()([x,y])

# 3 block (64-32):
y = tf.keras.layers.Conv2D(filters=512,kernel_size = (7,7),strides=(2, 2),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (5,5),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (5,5),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-6')(x)
x = tf.keras.layers.Conv2D(filters=512,kernel_size = (5,5),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Conv2D(filters=512,kernel_size = (5,5),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Add()([x,y])

# 4 block (32-16):
y = tf.keras.layers.Conv2D(filters=512,kernel_size = (5,5),strides=(2, 2),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (3,3),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (3,3),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-8')(x)
x = tf.keras.layers.Conv2D(filters=512,kernel_size = (3,3),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Conv2D(filters=512,kernel_size = (3,3),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Add()([x,y])

# 5 block (16-8):
y = tf.keras.layers.Conv2D(filters=1024,kernel_size = (3,3),strides=(2, 2),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (3,3),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.DepthwiseConv2D(kernel_size = (3,3),strides=1,padding='same',activation=None,activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.LeakyReLU()(x)
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=2,padding='valid',data_format=None,name='AveragePooling2D-10')(x)
x = tf.keras.layers.Conv2D(filters=1024,kernel_size = (3,3),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Conv2D(filters=1024,kernel_size = (3,3),strides=(1, 1),padding='same',activation=None,kernel_regularizer=tf.keras.regularizers.L2(),activity_regularizer=tf.keras.regularizers.L2())(x)
x = tf.keras.layers.Add()([x,y])


# 6 pooling 16->8:
x = tf.keras.layers.AveragePooling2D(pool_size=(2, 2),strides=1,padding='valid',data_format=None,name='AveragePooling2D')(x)
x = tf.keras.layers.BatchNormalization()(x)


x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(units=1024, kernel_initializer = 'uniform', activation='leaky_relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputed = tf.keras.layers.Dense(units=51,activation='softmax',activity_regularizer=tf.keras.regularizers.L2(), name='Output_layer')(x)


model_supclass = tf.keras.Model(inputed,outputed)

model_supclass.summary()

```
